I have sets of results for my cantica wrt all three metrics, and now I want to compare them.
Practically I'm comparing real sets of varying length (since the cantica have varying line-lengths). There are several options:

- simple: mean and standard deviation: measures how uniform the comp is
- dimension-reduction machine clustering on the sets direcly 
- polynomial regression: measures the shape; 
  - e.g. songs like nu01 would have a U shape and more or less fit to some kind of parabola.
- Wasserstein distance: difference both in location and shape

## Wasserstein distance

We start with the list `[[[[0.6666666666666666, 0.6666666666666666, ...]]]]` which is a superlist of play lists of canticum lists of line lists

```python
from comp import compatibility_corpus

all_sets = compatibility_corpus('compiled/')
print(all_sets)
```

```python
import numpy as np
from scipy.stats import wasserstein_distance

# Two sets of compatibility scores
set1 = np.array([0.8, 0.75, 0.9, 0.85, 0.78])
set2 = np.array([0.6, 0.65, 0.62, 0.7])

# Compute Wasserstein distance
distance = wasserstein_distance(set1, set2)

print(f"Wasserstein distance: {distance:.4f}")
```
